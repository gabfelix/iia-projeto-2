```python
%cd /content/darknet

import cv2
import darknet
import numpy

from google.colab.patches import cv2_imshow as colab_imshow

def image_detection(image, network, class_names, class_colors, thresh):
  # NOTE: Darknet doesn't accept numpy images
  # We create one with image we reuse for each detection
  width = darknet.network_width(network)
  height = darknet.network_height(network)
  darknet_image = darknet.make_image(width, height, 3)

  if isinstance(image, str):
    image = cv2.imread(image)

  try:
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  except:
    print(image)
    raise Exception('oy cvtColor caught an empty image again')

  image_resized = cv2.resize(image_rgb, (width, height), interpolation=cv2.INTER_CUBIC)

  darknet.copy_image_from_bytes(darknet_image, image_resized.tobytes())
  detections = darknet.detect_image(network, class_names, darknet_image, thresh=thresh)
  darknet.free_image(darknet_image)
  image = darknet.draw_boxes(detections, image_resized, class_colors)
  return cv2.cvtColor(image, cv2.COLOR_BGR2RGB), detections

def crop_detections_from_image(image, detections):
  if isinstance(image, str):
    image = cv2.imread(image)

  detected_object_images = []
  for detection in detections:
    w = int(detection[2][2])
    h = int(detection[2][3])

    x = int(detection[2][0] - (w / 2))
    y = int(detection[2][1] - (h / 2))

    label = detection[0]
    confidence = detection[1]
    cropped_img = image[int(y):int(y+h), int(x):int(x+w)]

    detected_object_images.append((label, confidence, cropped_img, (x, y, w, h)))
  return detected_object_images

def detect_objects_and_crop(image, network, class_names, class_colors, thresh):
  if isinstance(image, str):
    image = cv2.imread(image)

  if image.size == 0:
    raise Exception("detect_objects_and_crop(): empty image")

  result_image, detections = image_detection(image, network, class_names, class_colors, thresh)
  detection_images = crop_detections_from_image(cv2.resize(image, (darknet.network_width(network), darknet.network_height(network)), interpolation=cv2.INTER_LINEAR), detections)
  return detection_images

def draw_bounding_box(original_image, network, detections, label='', color=None):
  if isinstance(original_image, str):
    original_image = cv2.imread(original_image)

  drawn_image = original_image.copy() # FIXME: This copy might hurt performance
  for detection in detections:
    px, py, pw, ph = get_proportional_bounding_box_dimensions(original_image, detection, network)
    if color == None:
      color = (255, 0, 255)
    cv2.rectangle(drawn_image, (px, py), (px + pw, py + ph), color, 2)
  return drawn_image

def get_proportional_bounding_box_dimensions(original_image, detection, network):
  original_width, original_height = (original_image.shape[1], original_image.shape[0])
  network_width, network_height = (darknet.network_width(network), darknet.network_height(network))
  vehicle_x, vehicle_y, vehicle_w, vehicle_h = detection[3]
  prop_vx = int((vehicle_x / network_width) * original_width)
  prop_vy = int((vehicle_y / network_height) * original_height)
  prop_vw = int((vehicle_w / network_width) * original_width)
  prop_vh = int((vehicle_h / network_height) * original_height)
  return prop_vx, prop_vy, prop_vw, prop_vh

def full_detection_and_draw_image(original_image,
                                  vd_network = None, vd_class_names = None, vd_class_colors = None,
                                  lpd_network = None, lpd_class_names = None, lpd_class_colors = None,
                                  lpr_network = None, lpr_class_names = None, lpr_class_colors = None):
  # vd = vehicle detection
  VD_CONFIG='/content/darknet/vehicle-detection.cfg'
  VD_DATA='/content/darknet/vehicle-detection.data'
  VD_WEIGHTS='/content/darknet/vehicle-detection.weights'
  VD_THRESHOLD=0.25

  # lp = license plate
  # lpd = lp detection
  LPD_CONFIG='/content/darknet/lp-detection-layout-classification.cfg'
  LPD_DATA='/content/darknet/lp-detection-layout-classification.data'
  LPD_WEIGHTS='/content/darknet/lp-detection-layout-classification.weights'
  LPD_THRESHOLD=0.01

  # lpr = lp recognition
  LPR_CONFIG='/content/darknet/lp-recognition.cfg'
  LPR_DATA='/content/darknet/lp-recognition.data'
  LPR_WEIGHTS='/content/darknet/lp-recognition.weights'
  LPR_THRESHOLD=0.5

  # Define vehicle detection network
  if vd_network is None:
    vd_network, vd_class_names, vd_class_colors = darknet.load_network(
        VD_CONFIG,
        VD_DATA,
        VD_WEIGHTS
    )

  # Define LP detection network
  if lpd_network is None:
    lpd_network, lpd_class_names, lpd_class_colors = darknet.load_network(
        LPD_CONFIG,
        LPD_DATA,
        LPD_WEIGHTS
    )

  # Define LP recognition (OCR) network
  if lpr_network is None:
    lpr_network, lpr_class_names, lpr_class_colors = darknet.load_network(
        LPR_CONFIG,
        LPR_DATA,
        LPR_WEIGHTS
    )

  plates_contents = [] # Make sure we draw the plate contents last
  original_image_with_bb = original_image

  if isinstance(original_image, str):
    original_image = cv2.imread(original_image)

  if original_image.size == 0:
    raise Exception('ERROR: original_image is 0-sized')
  detected_vehicles = detect_objects_and_crop(original_image, vd_network, vd_class_names, vd_class_colors, VD_THRESHOLD)


  for cropped_vehicle in detected_vehicles:
    vehicle_label = cropped_vehicle[0]
    vehicle_confidence = cropped_vehicle[1]
    vehicle_image = cropped_vehicle[2]
    vehicle_bb_coordinates = cropped_vehicle[3]

    # Translate vehicle crop coordinates into original image coordinates
    vpx, vpy, vpw, vph = get_proportional_bounding_box_dimensions(original_image, ('', '', '', vehicle_bb_coordinates), vd_network)
    original_image_with_bb = cv2.rectangle(original_image, (vpx, vpy), (vpx + vpw, vpy + vph), (0, 0, 255), 2) # Draw vehicle bounding box

    # License Plate detection and classfication stage
    if vehicle_image.size == 0:
      print('WARN: vehicle_image is 0-sized')
      return original_image
    detected_lps = detect_objects_and_crop(vehicle_image, lpd_network, lpd_class_names, lpd_class_colors, LPD_THRESHOLD)

    if len(detected_lps) > 0:
      # Only use detected layout with highest confidence
      # FIXME: What if I detect a vehicle but no license plate inside it?
      highest_layout_confidence = 0.00
      highest_confidence_lp = None
      highest_layout_confidence = max([float(c[1]) for c in detected_lps])
      for cropped_lp in detected_lps:
        lp_confidence = float(cropped_lp[1])
        if lp_confidence == highest_layout_confidence:
          highest_confidence_lp = cropped_lp

      if (highest_confidence_lp == None):
        return original_image_with_bb

      lp_label = highest_confidence_lp[0]
      lp_confidence = highest_confidence_lp[1]
      lp_image = highest_confidence_lp[2]

      # Translate detection coordinates from license plate crop into original image coordinates
      lpx, lpy, lpw, lph = get_proportional_bounding_box_dimensions(vehicle_image, highest_confidence_lp, lpd_network) # plate crop dimensions -> vehicle crop dimensions
      lpx, lpy, lpw, lph = get_proportional_bounding_box_dimensions(original_image, ('','','', (lpx, lpy, lpw, lph)), vd_network) # vehicle crop dimensions -> original dimensions
      lpx += vpx
      lpy += vpy
      original_image_with_bb = cv2.rectangle(original_image_with_bb, (lpx, lpy), (lpx + lpw, lpy + lph), (0, 255, 0), 2) # Draw license plate bounding box

      # License Plate recognition (OCR) stage
      if lp_image.size == 0:
        print('WARN: lp_image was 0-sized') # FIXME
        return original_image_with_bb
      detected_characters = detect_objects_and_crop(lp_image, lpr_network, lpr_class_names, lpr_class_colors, LPR_THRESHOLD)
      # Sort detected LP characters by their X position
      chars = []
      for character in detected_characters:
        name = character[0]
        x = character[3][0]
        chars.append((name, x))
      chars = sorted(chars, key=lambda x: x[1])
      license_plate_contents = ''
      for char in chars:
        license_plate_contents += char[0]
      plates_contents.append((license_plate_contents, (lpx, lpy)))
      #original_image_with_bb = cv2.putText(original_image_with_bb, license_plate_contents, (int(lpx), int(lpy - 10)), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1.0, (0, 255, 0), 2, cv2.LINE_AA)
    for plate_content in plates_contents:
      plate_str = plate_content[0]
      lpx = int(plate_content[1][0])
      lpy = int(plate_content[1][1])
      original_image_with_bb = cv2.putText(original_image_with_bb, plate_str, (lpx, lpy - 10), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1.0, (0, 255, 0), 2, cv2.LINE_AA)

  return(original_image_with_bb) # Returns None if no detections

def full_detection_and_draw_video(video_path, output_path='/content/out.avi'):
  import time
  # vd = vehicle detection
  VD_CONFIG='/content/darknet/vehicle-detection.cfg'
  VD_DATA='/content/darknet/vehicle-detection.data'
  VD_WEIGHTS='/content/darknet/vehicle-detection.weights'
  VD_THRESHOLD=0.25

  # lp = license plate
  # lpd = lp detection
  LPD_CONFIG='/content/darknet/lp-detection-layout-classification.cfg'
  LPD_DATA='/content/darknet/lp-detection-layout-classification.data'
  LPD_WEIGHTS='/content/darknet/lp-detection-layout-classification.weights'
  LPD_THRESHOLD=0.01

  # lpr = lp recognition
  LPR_CONFIG='/content/darknet/lp-recognition.cfg'
  LPR_DATA='/content/darknet/lp-recognition.data'
  LPR_WEIGHTS='/content/darknet/lp-recognition.weights'
  LPR_THRESHOLD=0.5

  # Define vehicle detection network
  vd_network, vd_class_names, vd_class_colors = darknet.load_network(
      VD_CONFIG,
      VD_DATA,
      VD_WEIGHTS
  )

  # Define LP detection network
  lpd_network, lpd_class_names, lpd_class_colors = darknet.load_network(
      LPD_CONFIG,
      LPD_DATA,
      LPD_WEIGHTS
  )

  # Define LP recognition (OCR) network
  lpr_network, lpr_class_names, lpr_class_colors = darknet.load_network(
      LPR_CONFIG,
      LPR_DATA,
      LPR_WEIGHTS
  )

  # Read video frames
  vidcap = cv2.VideoCapture(video_path)
  success, image = vidcap.read()

  # Get video properties (frame width, height, and frame rate)
  frame_width = int(vidcap.get(cv2.CAP_PROP_FRAME_WIDTH))
  frame_height = int(vidcap.get(cv2.CAP_PROP_FRAME_HEIGHT))
  fps = int(vidcap.get(cv2.CAP_PROP_FPS))
  fourcc = cv2.VideoWriter_fourcc(*'XVID')  # avi
  out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

  timer = time.time()
  fixed_timer = 0.00
  count_last_second = 0
  count = 0
  fps = 0.00
  # Loop in every frame
  while success:
    if timer - fixed_timer >= 1:
      elapsed_time = timer - fixed_timer
      fps = float(count - count_last_second) / elapsed_time
      fixed_timer = timer
      count_last_second = count
    timer = time.time()

    print(f'processing frame #{count} (FPS: {fps:.2f})')


    new_frame = full_detection_and_draw_image(image,
                                  vd_network, vd_class_names, vd_class_colors,
                                  lpd_network, lpd_class_names, lpd_class_colors,
                                  lpr_network, lpr_class_names, lpr_class_colors)
    if new_frame is not None:
      out.write(new_frame)
    else:
      out.write(image)
    # Next frame
    count += 1
    success, image = vidcap.read()

  # Free resources
  vidcap.release()
  out.release()
  print(f'Finished processing. Output file: {output_path}')

# -- SCRIPT
from google.colab.files import download

SAMPLE_IMAGE='/content/darknet/sample-image.jpg'
colab_imshow(full_detection_and_draw_image(SAMPLE_IMAGE))
full_detection_and_draw_video('/content/video.mp4', '/content/out.avi')
download('/content/out.avi')
```
